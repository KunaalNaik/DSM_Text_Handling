```python
import re # ( regular expression)
```

```python
Text = "Prime Minister Narendra Modi  99i9i  + - :::::: , & ^ % () --- + 456on Sunday addressed India's 'gram sabhas' from Jammu's Palli village on the occasion National Panchayati Raj Diwas. PM Modi, on his first major visit to Jammu and Kashmir after the abrogation of Article 370 inaugurated and laid the foundation of multiple development initiatives worth Rs 20,000 crore in the Union Territory. "
```

```python
Text
```

```python
# Write down a regular expression to extract all small letter words. 
match = re.findall(r"[a-z]", Text)
print(match)

# Result - not the full words extracted but the small characters extracted 
```

```python
match = re.findall(r"[a-z]+", Text)
print(match)
```

```python
# Write a regular expression to extract both lower and uppercase 
match = re.findall(r"[a-zA-Z]+", Text)
print(match)
```

```python
# Write down regex to extract only digits 
match = re.findall(r"[0-9]+", Text)
print(match)
```

```python
# Regex to find everthing but not the numbers and not the characters 
match = re.findall(r"[^a-zA-Z0-9]+", Text)  # ^ is the NOT operator 
print(match)
```

```python
# Also skip space
match = re.findall(r"[^a-zA-Z0-9\s]+", Text)  # ^ is the NOT operator 
print(match)
```

```python
# Also skip dot 
match = re.findall(r"[^a-zA-Z0-9\s.]+", Text)  # ^ is the NOT operator 
print(match)
```

```python
# Also skip commas 
match = re.findall(r"[^a-zA-Z0-9\s.,]+", Text)  # ^ is the NOT operator 
print(match)
```

```python
# Task :- to remove all the special characters 
Temp1 = re.sub(r"[^a-zA-Z0-9\s.,]+"  , "", Text)
# re.sub( pattern, with what value you want to replace that patter, in which col/variable)
print(Temp1)
```

## How to translate text in the form on numbers 

```python
# Count Vectorization 
# Let's create a simple corpus with three documents 
corpus = [ 'This is sentence one', 'This is sentence two', 'This is sentence three']
```

```python
corpus
```

```python
type(corpus)
```

```python
from sklearn.feature_extraction import text
```

```python
cv = text.CountVectorizer(input = corpus )
```

```python
matrix = cv.fit_transform(corpus)
```

```python
matrix
```

```python
# Not recommended though - but just for viewing purpose 
# Translating this matrix in dataframe 
import pandas as pd 
pd.DataFrame(matrix.toarray(), columns= cv.get_feature_names())
```

```python
# Pair of words is called bi-grams 
cv1 = text.CountVectorizer(input = corpus,ngram_range=(2,2))
matrix1 = cv1.fit_transform(corpus)
pd.DataFrame(matrix1.toarray(), columns= cv1.get_feature_names())
```

```python
# create uni-gram and bi-gram both
cv1 = text.CountVectorizer(input = corpus,ngram_range=(1,2))
matrix1 = cv1.fit_transform(corpus)
pd.DataFrame(matrix1.toarray(), columns= cv1.get_feature_names())
```

```python
# Create tri-gram 
# create uni-gram and bi-gram both
cv1 = text.CountVectorizer(input = corpus,ngram_range=(3,3))
matrix1 = cv1.fit_transform(corpus)
pd.DataFrame(matrix1.toarray(), columns= cv1.get_feature_names())
```

```python
# Create unigram, bigram and tri-gram 
# create uni-gram and bi-gram both
cv1 = text.CountVectorizer(input = corpus,ngram_range=(1,3))
matrix1 = cv1.fit_transform(corpus)
pd.DataFrame(matrix1.toarray(), columns= cv1.get_feature_names())
```
